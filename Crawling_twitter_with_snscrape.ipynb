```python
import pandas as pd
import snscrape.modules.twitter as sntwitter

query = "(" + key + ")" + "lang:id until:" + until + "since:" + since
tweets = []
limit = 100000
save_interval = 1000
count = 0

print("Start crawling")
for tweet in sntwitter.TwitterSearchScraper(query=query).get_items():
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.content, tweet.hashtags])
        count += 1

    if count % save_interval == 0:
        df = pd.DataFrame(tweets, columns=['datetime', 'username', 'content', 'hashtags'])
        df.to_csv(f'tweets_{count}.csv', index=False)
        print(f"Saved {count} tweets")

df = pd.DataFrame(tweets, columns=['datetime', 'username', 'content', 'hashtags'])
df.to_csv('tweets.csv', index=False)

print("Finished")
print("--------")
```

```python
import csv
from googletrans import Translator

def translate_csv(input_file, output_file, column_index, source_lang, target_lang):
    translator = Translator()

    with open(input_file, 'r') as input_csv, open(output_file, 'w', newline='') as output_csv:
        reader = csv.reader(input_csv)
        writer = csv.writer(output_csv)

        for row in reader:
            if column_index < len(row):
                text = row[column_index]
                if text is not None and text.strip():  # Check for None values and empty strings
                    try:
                        translation = translator.translate(text, src=source_lang, dest=target_lang).text
                        row[column_index] = translation
                    except Exception as e:
                        print(f"Translation error for text '{text}': {str(e)}")
            writer.writerow(row)

# Example usage
input_file = '/content/Anies Baswedan.csv'
output_file = 'output_Anies Baswedan_3000_Data.csv'
column_index = 2  # Specify the column index (0-based) to translate
source_lang = 'en'  # Source language of the column
target_lang = 'id'  # Target language for translation

translate_csv(input_file, output_file, column_index, source_lang, target_lang)
```

```python
import csv
import os

def merge_csv_files(input_folder, output_file):
    csv_files = [file for file in os.listdir(input_folder) if file.endswith('.csv')]

    with open(output_file, 'w', newline='') as output_csv:
        writer = csv.writer(output_csv)

        for file in csv_files:
            file_path = os.path.join(input_folder, file)
            with open(file_path, 'r') as input_csv:
                reader = csv.reader(input_csv)
                for row in reader:
                    writer.writerow(row)

# Example usage
input_folder = '/content/dataset'  # Specify the folder path where the CSV files are located
output_file = 'merged_val_All.csv'  # Specify the output file name

merge_csv_files(input_folder, output_file)
```

```python
import csv

def count_rows(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        row_count = sum(1 for row in reader)
    return row_count

csv_file_path = '/content/merged_val_All.csv'
total_rows = count_rows(csv_file_path)
print("Total rows_Jijik_Muak:", total_rows)
```
